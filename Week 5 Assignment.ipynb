```python
import pandas as pd
import numpy as np
data = {
'Order_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
'Customer_Name': ['Amara', 'Bola', None, 'Diana', 'Efua', 'Fatima', 'Grace',
None, 'Ifeoma', 'Joyce'],
'Country': ['Ghana', 'Nigeria', 'Kenya', None, 'Ghana', 'Cameroon', None,
'Kenya', 'Nigeria', 'Ghana'],
'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', None, 'Phone', 'Tablet',
'Laptop', None, 'Phone'],
'Quantity': [2, 1, None, 3, 2, None, 1, 4, 2, None],
'Price': [1500, 800, 600, 1500, None, 800, 600, 1500, None, 800],
'Delivery_Days': [3, None, 7, 5, 3, None, 6, None, 4, 5]
}
df = pd.DataFrame(data)
print(df)

## Explore the Data
df.info()
print(df.info())

print(df.describe())
print(df)

df.isnull().sum()
print(df.isnull().sum())

# The data has 7 columns with missing values. The 'Customer_Name' column has 2 missing values, 'Country' has 2, 'Product' has 2, 'Quantity' has 3, 'Price' has 3, and 'Delivery_Days' has 4 missing values.
# The data types of the columns are as follows: 'Order_ID' is an integer, 'Customer_Name', 'Country', and 'Product' are objects (strings), while 'Quantity', 'Price', and 'Delivery_Days' are floats. The presence of missing values in these columns may affect data analysis and modeling, and appropriate handling of these missing values will be necessary to ensure accurate insights and predictions.


## Handling Missing Values

# Task 2 . Customer_Name

df['Customer_Name'].fillna('Unknown', inplace=True)
print(df)

# I have chosen to fill the missing values in the 'Customer_Name' column with 'Unknown' because it is a common practice to use a placeholder value for missing categorical data. This allows us to retain the structure of the dataset while acknowledging that the information is not available. Using 'Unknown' helps to differentiate between actual names and missing values, which can be useful for analysis and reporting.

# Task 3. Quantity, Delivery_Days, and Price


df['Quantity'].fillna(df['Quantity'].mean())
print(df)

# I have chosen to fill the missing values in the 'Quantity' column with the mean because it is a common method for handling missing numerical data. The mean provides a central value that can represent the overall distribution of the data. By filling in the missing values with the mean, we can maintain the integrity of the dataset and avoid introducing bias that may occur if we were to use other methods such as median or mode. This approach allows us to retain as much information as possible while minimizing the impact of missing values on our analysis.


df['Delivery_Days'] = df['Delivery_Days'].fillna(df['Delivery_Days'].median())
print(df)

df['Price'] = df['Price'].fillna(df['Price'].mean())
print(df)

# I have chosen to fill the missing values in the 'Delivery_Days' and 'Price' columns with the mean for the same reasons as mentioned for the 'Quantity' column. The mean provides a central value that can represent the overall distribution of the data, allowing us to maintain the integrity of the dataset while minimizing the impact of missing values on our analysis. By filling in these missing values with the mean, we can retain as much information as possible and ensure that our analysis is not skewed by missing data.

# For 'Country' and 'Product' columns, I have chosen to leave the missing values as they are because filling them with a placeholder value may not be appropriate in this context. The 'Country' and 'Product' columns contain categorical data, and filling missing values with a placeholder such as 'Unknown' may not provide meaningful insights for analysis. Instead, it may be more beneficial to analyze the data with the missing values intact or to consider other methods of handling missing categorical data, such as using the most frequent category or creating a separate category for missing values if necessary.
```
